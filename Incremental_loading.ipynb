## 5a. Incremental loading of table PatientInformation from az sql server to bronze table

```
%sql
MERGE INTO hive_metastore.incremental_lakehouse.PatientInformation AS target
USING (
    SELECT 
        patient_id,
        first_name,
        last_name,
        date_of_birth,
        gender,
        blood_type,
        medical_history,
        primary_physician,
        date_of_admission,
        date_of_discharge,
        diagnosis,
        treatment_plan
    FROM (
        SELECT
            patient_id,
            first_name,
            last_name,
            date_of_birth,
            gender,
            blood_type,
            medical_history,
            primary_physician,
            date_of_admission,
            date_of_discharge,
            diagnosis,
            treatment_plan,
            ROW_NUMBER() OVER (PARTITION BY patient_id ORDER BY date_of_admission DESC) AS row_num
        FROM PatientInformation_df
    ) AS deduplicated_source
    WHERE row_num = 1
) AS source
ON target.patient_id = source.patient_id

WHEN MATCHED THEN 
UPDATE 
SET 
    target.first_name = source.first_name,
    target.last_name = source.last_name,
    target.date_of_birth = source.date_of_birth,
    target.gender = source.gender,
    target.blood_type = source.blood_type,
    target.medical_history = source.medical_history,
    target.primary_physician = source.primary_physician,
    target.date_of_admission = source.date_of_admission,
    target.date_of_discharge = source.date_of_discharge,
    target.diagnosis = source.diagnosis,
    target.treatment_plan = source.treatment_plan

WHEN NOT MATCHED THEN 
INSERT (
    patient_id,
    first_name,
    last_name,
    date_of_birth,
    gender,
    blood_type,
    medical_history,
    primary_physician,
    date_of_admission,
    date_of_discharge,
    diagnosis,
    treatment_plan
) 
VALUES (
    source.patient_id,
    source.first_name,
    source.last_name,
    source.date_of_birth,
    source.gender,
    source.blood_type,
    source.medical_history,
    source.primary_physician,
    source.date_of_admission,
    source.date_of_discharge,
    source.diagnosis,
    source.treatment_plan
)

WHEN NOT MATCHED BY SOURCE THEN 
DELETE;
```

### Create a function to rename column and to create md5 hash column based on SCD cloumns

```
def column_renamer(df, suffix, append):
   
    if append:
        new_column_names = list(map(lambda x: x+suffix, df.columns))
        
    else:
        new_column_names = list(map(lambda x: x.replace(suffix,""), df.columns))
        
    return df.toDF(*new_column_names)

def get_hash(df, keys_list):
 
    columns = [col(column) for column in keys_list]
    
    if columns:
        return df.withColumn("hash_md5", md5(concat_ws("", *columns)))
    else:
        return df.withColumn("hash_md5", md5(lit(1)))
```

```
from pyspark.sql.functions import *
from pyspark.sql.window import Window
DATE_FORMAT = "yyyy-MM-dd"
future_date = "9999-12-31"
window_spec = Window.orderBy("patient_id")
slowly_changing_cols = ['contact_id', 'patient_id', 'email', 'address_line1', 'address_line2', 'city', 'state', 'postal_code', 'emergency_contact_phone', 'emergency_contact_relationship']

target_df = target_df \
.withColumn("surrogate_key",row_number().over(window_spec)) \
.withColumn("effective_date",date_format(current_date(), DATE_FORMAT)) \
.withColumn("end_date",date_format(lit(future_date), DATE_FORMAT)) \
.withColumn("active_flag", lit(True))
```

```
merged_df = active_target_rename_df.join(source_df_rename_df, col("contact_id_target") == col("contact_id_source"), "full_outer") \
.withColumn("hash_md5_source", col("hash_md5_source").cast("string")) \
.withColumn("hash_md5_target", col("hash_md5_target").cast("string")) \
.withColumn("Action", when(col("hash_md5_source") == col("hash_md5_target"), 'NOCHANGE')
.when(col("patient_id_source").isNull(), 'DELETE')
.when(col("patient_id_target").isNull(), 'INSERT')
.otherwise('UPDATE'))
display(merged_df)
```

#### INCREMNETALLY LOAD THE DATA INTO incremental_lakehouse.PatientContactDetails_incrementalLoad (FINAL TABLE)

```
from pyspark.sql import functions as F
from pyspark.sql.functions import col
from delta.tables import DeltaTable

# Define the source DataFrame (Finalised_df_2)

# Define the target Delta table
target_table = "incremental_lakehouse.PatientContactDetails_incrementalLoad"

# List of columns that match between Finalised_df_2 and the target table
matching_columns = [
    "contact_id", "patient_id", "phone_number", "email", "address_line1", "address_line2",
    "city", "state", "postal_code", "emergency_contact_name", "emergency_contact_phone",
    "emergency_contact_relationship"
]

# Create a DeltaTable object for the target table
deltaTable = DeltaTable.forName(spark, target_table)

### Handle INSERT operation (select only the matching columns)
insert_df = Finalised_df_2.filter((col("Action") == "INSERT")).select(*matching_columns)

# Perform INSERT operation
if insert_df.rdd.isEmpty() == False:
    insert_df.write.mode("append").format("delta").saveAsTable(target_table)


### Handle UPDATE operation (using Delta MERGE for efficient updates)
# Prepare update DataFrame by filtering "UPDATE" action and selecting distinct contact_id
update_df = Finalised_df_2.filter((col("Action") == "UPDATE") & (col("active_flag") == "true")) \
                            .select(*matching_columns)

# Perform MERGE operation for updates
if update_df.rdd.isEmpty() == False:
    # Merge to update existing records based on contact_id
    deltaTable.alias("tgt").merge(
        update_df.alias("src"),
        "tgt.contact_id = src.contact_id"
    ).whenMatchedUpdate(set={
        f"tgt.{col}": f"src.{col}" for col in matching_columns
    }).execute()


### Handle DELETE operation (using Delta MERGE for efficient deletes)
# Prepare delete DataFrame by filtering "DELETE" action and selecting only contact_id
delete_df = Finalised_df_2.filter((col("Action") == "DELETE")).select("contact_id")

# Perform MERGE operation for deletes
if delete_df.rdd.isEmpty() == False:
    # Merge to delete records where contact_id matches
    deltaTable.alias("tgt").merge(
        delete_df.alias("src"),
        "tgt.contact_id = src.contact_id"
    ).whenMatchedDelete().execute()
```

```
display(spark.sql("describe history incremental_lakehouse.PatientContactDetails_incrementalLoad"))
# Step 1: Get the history of the PatientInformation table
history_df = spark.sql("DESCRIBE HISTORY incremental_lakehouse.PatientContactDetails_incrementalLoad")

# Step 2: Select the two most recent versions
recent_versions = history_df.select("version").orderBy("version", ascending=False).limit(2).collect()

# Extract the version numbers
version_1 = recent_versions[0][0]  # Most recent version
version_2 = recent_versions[1][0]  # Second most recent version
print(version_1, version_2)
# Step 3: Create DataFrames using the two most recent versions
df1a = spark.sql(f"SELECT * FROM table_changes('incremental_lakehouse.PatientContactDetails_incrementalLoad', {version_1})")
df2a = spark.sql(f"SELECT * FROM table_changes('incremental_lakehouse.PatientContactDetails_incrementalLoad', {version_2})")


# Find the difference between the two DataFrames
diff_df1a = df2a.subtract(df1a) #old_records

# Display the differences
display(diff_df1a)
```
